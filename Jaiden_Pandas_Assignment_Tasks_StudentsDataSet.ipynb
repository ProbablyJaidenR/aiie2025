```python
#Replace missing Grade values with the column's mean.
df['Grade'].fillna(df['Grade'].mean(), inplace=True)
```

```python
#David went from null to the average
df.iloc[:23]
```

```python
#Fill missing Attendance values with a constant value (e.g., 0).
df.fillna({'Attendance':0}, inplace=True)
```

```python
#Charlie went from null to 0
df.iloc[:1]
```

```python
#Drop any rows where the Name or StudentID is missing.
df.dropna(subset=['Name', 'StudentID'], inplace=True)
#Grace and Mia are removed
df.iloc[60:65]
```

```python
#Standardize the Gender column to ensure values are consistent (e.g., replace "M" with "Male" and "F" with "Female").
df['Gender'].replace({'M':'Male', 'F': 'Female', 'male': 'Male', 'female' : 'Female'}, inplace=True)
df.iloc[:23]
```

```python
def pass_fail(grade):
  if grade >= 50:
    return 'Pass'
  else:
    return 'Fail'
```

```python
#Add a new column called Pass/Fail:
#A student passes if their Grade is 50 or above.

df['Pass/Fail'] = df['Grade'].apply(pass_fail)
df
```

```python
#Create a new DataFrame containing only students who have Attendance greater than 75%.

attendance_df = df[df['Attendance'] > 75]

#Sort the DataFrame by Grade in descending order.
attendance_df.sort_values(by='Attendance', ascending=False, inplace=True)
attendance_df
```

```python
#Group the data by Subject and calculate the average Grade for each subject.

df_subject = df.groupby('Subject').describe()
df_subject
```

```python
#Find the top 3 students with the highest grades in each subject.

top_three = df.groupby('Subject').apply(lambda x: x.nlargest(3,'Grade'))
top_three
```

```python
#Create a pivot table showing the average Grade of students for each Subject
#and Gender.

df.pivot_table(index= 'Subject',
               columns='Gender',
               values='Grade',
               aggfunc='mean')
```
